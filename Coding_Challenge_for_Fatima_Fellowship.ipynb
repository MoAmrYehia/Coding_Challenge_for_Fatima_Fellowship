```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```
dataset_true = pd.read_csv("drive/MyDrive/FakeNews/True.csv")
dataset_true['true'] = 1
dataset_true
```

```
dataset_fake = pd.read_csv("drive/MyDrive/FakeNews/Fake.csv")
dataset_fake['true'] = 0
dataset_fake
```

```
full_news = pd.concat([dataset_true,dataset_fake])
full_news
```

```
target = ['true']
data = ['text']

target_column = full_news[target]
text_column = full_news[data]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(text_column, target_column , test_size=0.2, random_state=42)
```

```
import ktrain
from ktrain import text
```

```
max_len = 500
batch_size = 6
learning_rate = 5e-5
epochs = 1
```

```
model_ = 'bert-base-uncased'
text_model = text.Transformer(model_, maxlen=500, classes = [0,1])

#train
x_train = x_train['text'].tolist()
y_train = y_train['true'].tolist()

#test
x_test = x_test['text'].tolist()
y_test = y_test['true'].tolist()


# Pre-processing training & test data
train = text_model.preprocess_train(x_train,y_train)
test = text_model.preprocess_train(x_test,y_test)

# Model Classifier
model = text_model.get_classifier()

learner = ktrain.get_learner(model, train_data=train, val_data=test, batch_size=6)
```

```
learner.fit_onecycle(learning_rate, epochs)
```

```
x = learner.validate(class_names=text_model.get_classes())
```

```
predictor = ktrain.get_predictor(learner.model, preproc = text_model)
predictor.save('drive/MyDrive/FakeNews/model')
```
